```python
multirun_evaluators  = {}
_dir = 'avg'

lang_codes_to_evaluate = ['en', 'it']
lang_pair = ['en', 'it']

for dimensions in sorted([int(i) for i in diff_dims]):
    r = 1
    c = 2
    fig, axes = plt.subplots(1, 2, figsize=(20,7), sharey=True)
    fig.suptitle("Language pair {}, Dimensions {} - {}".format('_'.join(lang_pair), dimensions, _dir))
    
    for i, eig_tol in enumerate(sorted(diff_eigs)):
        matched_runs_dims = utils.filter_runs_on_field(experiment_id, 'dimensions', dimensions, verbose = False)
        experiment_run_ids_dims = [run['id'] for run in matched_runs_dims]
        
        matched_runs_eigs = utils.filter_runs_on_field(experiment_id, 'eigs_tol_1', eig_tol, verbose = False)
        experiment_run_ids_eigs = [run['id'] for run in matched_runs_eigs]
        
        experiment_run_ids = set(experiment_run_ids_dims) & set(experiment_run_ids_eigs)
        
        assert(len(experiment_run_ids) != 0)
        
        mr_eval_obj = evaluator.Multirun_evaluator(experiment_run_ids, lang_codes_to_evaluate)
        
        ax = axes[i]
        
        ax.set_ylim([1.2, 2.2])
        ax.set_xlabel('Regularization parameter')
        ax.set_ylabel('Harmonic mean of ranks')
        mr_eval_obj.plot_performance(lang_pair, _dir, 'Eigs tol %s' % eig_tol, ax)
    
    plt.savefig('../plots/{}_dim_{}_{}.png'.format(experiment_id, dimensions, '_'.join(lang_pair)), bbox_inches='tight')
```
